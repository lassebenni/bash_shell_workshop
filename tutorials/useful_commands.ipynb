## Useful commands

#### Disk space usage

```bash
df -h
```

#### File space usage

```bash
du -h
```

#### Grep for a single word in dir

```bash
grep -rw threatened ~
```

#### shuf to randomly take items

```bash
shuf -n 2 names.txt
```

#### diff to compare files

```bash
cp names.txt names2.txt
```

```bash
diff -s names.txt names2.txt
```

```bash
diff names.txt names2.txt
```

#### Xargs to perform actions per item

```bash
cat names.txt
```

```bash
cat names.txt | xargs -I % echo "% is a person."
```

### sorting with sort

```bash
sort names.txt
```

### wget to download webpages

```bash
wget -kp klm.txt https://nl.wikipedia.org/wiki/Koninklijke_Luchtvaart_Maatschappij
```

```bash
cat klm.txt
```

```bash
wget -O erruptions.csv https://people.sc.fsu.edu/~jburkardt/data/csv/faithful.csv
```

```bash
cat erruptions.csv
```

```bash
sort -k 3 erruptions.csv 
```
